```python
def convert_votes(vote_str):
    try:
        vote_str = str(vote_str).strip("()")  
        if 'M' in vote_str:
            return int(float(vote_str.replace('M', '')) * 1000000)
        elif 'K' in vote_str:
            return int(float(vote_str.replace('K', '')) * 1000)
        else:
            return int(vote_str)
    except:
        return np.nan


dc['Votes'] = dc['Votes'].apply(convert_votes)
```

```python
#  Fill missing Votes with median
votes_median = dc['Votes'].median()
dc['Votes'] = dc['Votes'].fillna(votes_median)
```

```python
#Extract hours and minutes separately
dc["hours"] = dc["Duration"].str.extract(r'(\d+)h').astype(float)
dc["minutes"] = dc["Duration"].str.extract(r'(\d+)m').astype(float)

# Fill missing values with 0
dc['hours'] = dc['hours'].fillna(0)
dc['minutes'] = dc['minutes'].fillna(0)

#  Calculate total minutes
dc['total_minute'] = dc['hours'] * 60 + dc['minutes']

#  Convert total_minute to integer
dc['total_minute'] = dc['total_minute'].astype(int)

# Drop the helper columns
dc.drop(columns=['hours', 'minutes'], axis=1, inplace=True)

#  Drop the original Duration column
dc.drop(columns=["Duration"], axis=1, inplace=True)
```

```python
#  Fill missing Genres with 'Unknown'
dc['Genres'] = dc['Genres'].fillna('Unknown')

#  Create a function to get last 3 genres
def get_last_genres(genre_str):
    genres_list = genre_str.split(',')
    genres_list = [g.strip() for g in genres_list]  # remove spaces
    if len(genres_list) >= 3:
        return ', '.join(genres_list[-3:])  # take LAST 3
    else:
        return ', '.join(genres_list)  # take all available

#  Apply the function
dc['Main_Genre'] = dc['Genres'].apply(get_last_genres)
```

```python
genre_mapping = {
    'Action Epic': 'Action',
    'One-Person Army Action': 'Action',
    'Cop Drama': 'Crime',
    'Legal Drama': 'Drama',
    'Psychological Drama': 'Drama',
    'Political Drama': 'Drama',
    'Docudrama': 'Drama',
    'Coming-of-Age': 'Drama',
    'Martial Arts': 'Action',
    'Romantic Comedy': 'Comedy',
    'Quirky Comedy': 'Comedy',
    'Buddy Comedy': 'Comedy',
    'Dark Comedy': 'Comedy',
    'Slapstick': 'Comedy',
    'Showbiz Drama': 'Drama',
    'Folk Horror': 'Horror',
    'Supernatural Horror': 'Horror',
    'Psychological Thriller': 'Thriller',
    'Conspiracy Thriller': 'Thriller',
    'Political Thriller': 'Thriller',
    'Suspense Mystery': 'Mystery',
    'Quest': 'Adventure',
    'Adventure Epic': 'Adventure',
    'Historical Epic': 'History',
    'War Epic': 'War',
    'Medical Drama': 'Drama',
    'Crime Documentary': 'Crime',
    'Spy': 'Action',
    'Road Trip': 'Adventure',
    'Jungle Adventure': 'Adventure',
    'Artificial Intelligence': 'Sci-Fi',
    'Body Horror': 'Horror',
    'Serial Killer': 'Crime',
    'True Crime': 'Crime',
    'Parody': 'Comedy',
    'Heist': 'Crime',
    'Monster Horror': 'Horror',
    'Swashbuckler': 'Adventure'
}
```

```python
def clean_main_genre(genre_list):
    if pd.isnull(genre_list):
        return 'Unknown'
    parts = genre_list.split(',')
    cleaned = []
    for part in parts:
        part = part.strip()
        if part in genre_mapping:
            cleaned.append(genre_mapping[part])
        else:
            cleaned.append(part)
    return ', '.join(cleaned)

    
dc['Genre_Cleaned'] = dc['Main_Genre'].apply(clean_main_genre)
```

```python
print(dc['Genre_Cleaned'].head(20))
```

```python
def remove_duplicates(genre_string):
    if pd.isnull(genre_string):
        return 'Unknown'
    parts = genre_string.split(',')
    parts = [part.strip() for part in parts]
    parts = list(dict.fromkeys(parts))  # Remove duplicates and keep order
    return ', '.join(parts)

# Apply to Genre_Cleaned
dc['Genre_Cleaned'] = dc['Genre_Cleaned'].apply(remove_duplicates)
```

```python
#  Fill missing Gross Worldwide using Genre_Cleaned group median
dc['Gross Worldwide'] = dc['Gross Worldwide'].fillna(
    dc.groupby('Genre_Cleaned')['Gross Worldwide'].transform('median')
)

#  If still missing, fill using overall median
dc['Gross Worldwide'] = dc['Gross Worldwide'].fillna(
    dc['Gross Worldwide'].median()
)
```

```python
#Why?

#Step	Why it is good
#Fill by Genre Median	- Different Genres earn differently (Action movies usually earn more than Art movies) — so Genre-wise median is more logical
#Fill remaining with Overall Median	- If Genre data is not available, then overall median gives a safe fallback
```
